```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df = pd.read_csv('activities.csv')
```

```python

df['Activity Date'] = pd.to_datetime(df['Activity Date'], format='%d %b %Y, %H:%M:%S')
start_date = pd.to_datetime('2020-01-01')
df = df[df['Activity Date'] >= start_date]
df.set_index('Activity Date', inplace=True)
monthly_data = df['Distance'].resample('M').mean()
plt.figure(figsize=(10, 6))
plt.plot(monthly_data.index,monthly_data.values,linestyle='-')
plt.xlabel('Activity Date')
plt.ylabel('Average Distance (KM)')
plt.title('Running Distance from 2020-2023')
plt.grid(True)

num_months = 2
locator = mdates.MonthLocator(interval=num_months)
plt.gca().xaxis.set_major_locator(locator)
plt.xticks(rotation=45) 
plt.show()


```

This is my personal running data. Back when COVID-19, I started running due to the gyms being closed. I started tracking my runs once I finished the running program "Couch to 5k". After that I started the second program "Couch to 10k". Which you can see in the graph in October 2020, my average distance jumped from 5km to 10km. After finishing the 10k program, I wanted to find another program that will help me reach my goal of running a marathon.But with work, I didn't have the time to commit more than hour of running. So I tried to just run around 10 km whenever I had the time.

As you can see around June 2022, I went back to school so I had some more time to focus on running. So I decided to add an extra 1km after a month of running. I stopped at 13km because school was starting, the weather was getting cold and I had no interest staying outside to just run an extra kilometer. 

Currently I'm stuck at the 13 kilometer plateau, I'm not certain when I will be breaking out that area soon.



```python
monthly_runs = df['Distance'].resample('M').count()
plt.figure(figsize=(10, 6))
plt.plot(monthly_data.index,monthly_runs.values,linestyle='-')
plt.xlabel('Activity Date')
plt.ylabel('Number of Runs')
plt.title('Number of Runs 2020-2023')
plt.grid(True)

num_months = 2
locator = mdates.MonthLocator(interval=num_months)
plt.gca().xaxis.set_major_locator(locator)
plt.xticks(rotation=45) 
plt.show()
```

When I first started the Couch to 10k program, I was eager to quickly finish it. Which is why at the beginning my number of runs was really high because I was running almost every single day. But once I finished, I started running every other day. Additionally you notice big dips around January-March becuase it's winter time and it is really hard to run in the snow.
